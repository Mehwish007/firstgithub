# import libraries
seaborn automatically installs these libraries
- numpy

- pandas
- matplotlib

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool=sns.load_dataset("iris")
phool
#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
#sepal=petal of flower
plt.show()
```

### adding title

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool=sns.load_dataset("iris")
phool
#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("phoolon ka plot")
#sepal=petal of flower
plt.show()
```

### adding x and y limits

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool=sns.load_dataset("iris")
phool
#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("phoolon ka plot")
plt.xlim(2)
plt.ylim(1)
#sepal=petal of flower
plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool=sns.load_dataset("iris")
phool
#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("phoolon ka plot")
sns.set_style("dark")
sns.set_style(style=None, rc=None)
#sepal=petal of flower
plt.show()
```

## size of figure

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool=sns.load_dataset("iris")
# change figure
plt.figure(figsize=(4,2))
#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("phoolon ka plot")


#sepal=petal of flower
plt.show()
```
